```python
# imports

import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
```

```python
os.getcwd()
```

```python
sample_200 = pd.read_csv('../files/datasets/labeled/l01_reuters_sample200.csv')
sample_800 = pd.read_csv('../files/datasets/labeled/l02_reuters_sample800.csv')
```

```python
sample_800
```

```python
labeled_sentences = pd.concat([sample_200, sample_800])
```

```python
labeled_sentences.drop('Unnamed: 0', axis=1)
```

```python
# make plots over the valence/arousal distributions

fig = plt.figure(figsize=(20,10))
gs = fig.add_gridspec(nrows=1, ncols=2)
axs = gs.subplots()
print(axs[0])

sns.distplot(labeled_sentences['valence'], ax=axs[0])
axs[0].grid()
sns.distplot(labeled_sentences['arousal'], ax=axs[1])
axs[1].grid()

plt.savefig('../files/results/graphs/va_dist.png')
```

```python
# calculate normality-distribution p-value and make distribution summary stats

valence_pval = shapiro(labeled_sentences['valence'])
arousal_pval = shapiro(labeled_sentences['arousal'])
summ_stats = labeled_sentences[['valence', 'arousal']].describe()
summ_stats = summ_stats.drop(index=['count'])
summ_stats.loc['p-val'] = [round(valence_pval[1], 8), round(arousal_pval[1], 11)] 
summ_stats = summ_stats.round(decimals=2)

print(round(valence_pval[1], 8))
print(round(arousal_pval[1], 11))
print(summ_stats.to_latex())
```
