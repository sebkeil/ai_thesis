```python
# general imports
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read in joint training logs for valence and arousal

import pandas as pd

def parse_file(file_name):
    """
    Takes in a file name, e.g., v_joint_train.txt, parses its content and converts it to a pandas dataframe.
    
    :param file_name str: the name of the txt file
    
    :return res_df pandas DataFrame object: a parsed dataframe containing the cleaned results
    """
    file_type = file_name.split("_")[-1]
    
    joint_train_file = results_folder + file_name

    with open(joint_train_file, "r") as file:
        raw_text = file.read()

    res = {}
    
    if file_type == "train.txt":
        epochs = raw_text.split("\n")[:-1]
        for epoch in epochs:
            parsed_epoch_str = epoch.split("|")
            name, rmse_list = parsed_epoch_str[0], parsed_epoch_str[1]

            # parse the rmse_list, which is now just a giant string of results
            rmse_list = rmse_list[1:-1]
            rmse_list = rmse_list.split(",")

            rmse_list = [float(entry) for entry in rmse_list]

            res[name] = rmse_list
            
    if file_type == "test.txt":
        rmse_list = raw_text[1:-1].split(",")
        rmse_list = [float(entry) for entry in rmse_list]
        res['test'] = rmse_list
            
    res_df = pd.DataFrame(res)
    return res_df
```

```python
v_joint_train = parse_file("v_joint_train.txt")
```

```python
v_disjoint_train = parse_file("v_disjoint_train.txt")
```

```python
# make plot of valence TRAINING loss (joint vs. disjoint) for each epoch
# purpose: analyze training behavior

fig = plt.figure(figsize=(20,15))
gs = fig.add_gridspec(nrows=3, ncols=1)
axs = gs.subplots()

for i, col in enumerate(v_joint_train.columns):
    axs[i].plot(v_joint_train[col], label='Joint loss')
    axs[i].plot(v_disjoint_train[col], label='Disjoint loss')
    axs[i].set_title(f'Epoch {i+1}')
    axs[i].grid()
    axs[i].legend()
    
plt.xlabel('Iteration/Epoch')
plt.ylabel('RMSE Loss')
plt.savefig(os.getcwd() + '\\files\\results\\graphs\\v_joint_disjoint_training.png')
plt.show()
```

```python
a_joint_train = parse_file("a_joint_train.txt")
a_disjoint_train = parse_file("a_disjoint_train.txt")
```

```python
# make plot of arousal TRAINING loss (joint vs. disjoint) for each epoch
# purpose: analyze training behavior

fig = plt.figure(figsize=(20,15))
gs = fig.add_gridspec(nrows=3, ncols=1)
axs = gs.subplots()

for i, col in enumerate(a_joint_train.columns):
    axs[i].plot(a_joint_train[col], label='Joint loss')
    axs[i].plot(a_disjoint_train[col], label='Disjoint loss')
    axs[i].set_title(f'Epoch {i+1}')
    axs[i].grid()
    axs[i].legend()
    
plt.xlabel('Iteration/Epoch')
plt.ylabel('RMSE Loss')
plt.savefig(os.getcwd() + '\\files\\results\\graphs\\a_joint_disjoint_training.png')
plt.show()
```

```python
# load in test data
v_joint_test = parse_file("v_joint_test.txt")
a_joint_test = parse_file("a_joint_test.txt")
v_disjoint_test = parse_file("v_disjoint_test.txt")
a_disjoint_test = parse_file("a_disjoint_test.txt")
```

```python
def calculate_training_mean(df):
    e1 = np.mean(df['epoch_1'])
    e2 = np.mean(df['epoch_2'])
    e3 = np.mean(df['epoch_3'])
    return (e1+e2+e3)/3
```

```python
# make a table comparing all the values

res = {}

res['joint'] = {}
res['disjoint'] = {}
res['joint']['valence'] = {}
res['joint']['arousal'] = {}
res['disjoint']['valence'] = {}
res['disjoint']['arousal'] = {}

res['joint']['valence']['test'] = round(np.mean(v_joint_test['test']),3)
res['joint']['valence']['train'] = round(calculate_training_mean(v_joint_train),3)
res['joint']['arousal']['test'] = round(np.mean(a_joint_test['test']),3)
res['joint']['arousal']['train'] = round(calculate_training_mean(a_joint_train),3)

res['disjoint']['valence']['test'] = round(np.mean(v_disjoint_test['test']),3)
res['disjoint']['valence']['train'] = round(calculate_training_mean(v_disjoint_train),3)
res['disjoint']['arousal']['test'] = round(np.mean(a_disjoint_test['test']),3)
res['disjoint']['arousal']['train'] = round(calculate_training_mean(a_disjoint_train),3)

res_df = pd.DataFrame(res)



```
